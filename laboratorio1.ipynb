##**Git / Github**##

Git  es un sistema de control de versiones de archivos y usa una terminología similar para archivos y fólderes.  Lo que es llamado un folder en git es llamado un *tree*. Lo que es llamado un archivo en git es llamado *blob*. 
Git usa un grafo para mantener track de lo que sucede. 
Git incorpora snapshots de  cada estado de los archivos y los fólderes para mantener una historia.
Detrás de cada snapshot existe metadata para describir un estado o snapshot.

>Cada snapshoot es un `commit`. Un commit es un conjunto de padres referenciados o Ids. Los padres son varios `commits`. Todos estos son objetos.


Durante la lección, el profesor usa los comandos más importantes para el manejo profesional de esta herramienta. Las cuales se listan a continuación.




**Comandos usados en la lección**


* `mkdir <name>`  Este comando  crea una carpeta llamada con el nombre que se le especifica, en lugar actual dentro de la consola.

* `cd <name>` Este comando ingresa a la carpeta que se le especifica.

* `ls`     Este comando usado para listar todos los archivos y carpetas dentro de del directorio.


* `git init`     Crea un repositorio vacio dentro del directorio actual.

* `ls –a`   Este comando permite observar archivos dentro de la carpeta que estén ocultos.

* `git help init`  Provee información acerca git, especificamente de init.


* `git status` Permite ver el estado del repositorio. 

* `git add .`  agrega todos archivos del directorio  al repositorio para que sea controlados.

* `git commit`  Crea un snapshot de los archivos y las carpetas. Es  importante escribir mensajes con alta calidad.

La diferencia entre `add` y `commit` es que en ocasiones no puedes agregar todo al repositorio.

* `git log –all –graph –decorate`     Sirve para mostrar todo los cambios ocurridos en el repositorio de forma gráfica.

*Master* hace referencia a la versión más actualizada de tu proyecto.
*HEAD*  es usada para referenciar a donde estás viendo actualmente. 

* `git checkout <name of branch>` # Ayuda a moverte en las diferentes ramas.  Por ejemplo, para volver al master.   `git checkout master` me ayuda a mover el head a la rama principal.

* `Vim hello.txt`   sirve para abrir el archivo hello.txt.

Para cerrar el Vim  Tienes varias opciones, pulsar **Esc** y escribir el comando :q pero con exclamación cerrada, es decir: **Esc :q!** para salir sin guardar.

* `git diff` sirve para encontrar las diferencias entre dos commits distintos.
* `git branch  “Name”`   creará una nueva rama de git.
* `git merge “cat”`   combina la rama actual en master.





**Git Remoto**

En varias ocasiones es necesario trabajar con equipos.  Entonces, ante esa necesidad es necesario usar un repositorio remoto para trabajar con otras personas.

Los comandos más usados en estos casos, mencionados por el profesor, fueron:


1. `git remote`   Checa todos los repositorios remotos.
2. `git remote add origin ../remote`   Agrega un repositorio remoto a staging.
3. `git push origin master`   Envia los cambios al repositorio remoto.
4. `git clone  <url> <where>`  Clona un repositorio remoto en uno local.
5. `git fetch`   Obtiene los cambios del repositorio remoto.
6. `git pull` Obtiene los cambios en la maquina local desde el remoto
7. `git config`   Sirve para configurar tu cuenta en la que estas trabajando.
8. `git ignore`  Sirve para decirle a git que archivos ignorar, por ejemplo los binarios.
